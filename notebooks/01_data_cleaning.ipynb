```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/listings.csv")
```

```python
cols = [
    "id",
    "host_id",
    "neighbourhood_group_cleansed",
    "neighbourhood_cleansed",
    "latitude",
    "longitude",
    "room_type",
    "minimum_nights",
    "number_of_reviews",
    "availability_365",
    "calculated_host_listings_count"
]

df = df[cols].copy()
```
